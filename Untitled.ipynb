```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
ibm = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
ibm
```

```python
# checking for missing values
print(ibm.isnull().sum())
```

```python
ibm.head()
```

```python
ibm.describe()
```

```python
#All values have the same value and Irrelevant variable for the analysis
ibm.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)
```

```python
ibm.dtypes
```

```python
#categorisation of dtypes i.e object
ibm.Attrition = ibm.Attrition.astype('category').cat.codes
```

```python
categorical_features = []
for column in ibm.columns:
    if ibm[column].dtype == object:
        categorical_features.append(column)
        print(f"{column}")
        print("====================================")
```

```python
numerical_features = []
for column in ibm.columns:
    if ibm[column].dtype != object:
        numerical_features.append(column)
        print(f"{column}")
        print("====================================")
```

```python
numerical_features.remove('Attrition')
```

```python
ibm["Attrition"].value_counts()
```

```python
ibm.loc[ibm['Attrition'] == 'Yes','Attrition'] = '1'
ibm
```

```python
sns.countplot(x='Attrition', data=ibm)
plt.title('Attrition Distributions \n (0: No Attrition || 1: Attrition)')
```

```python
ibm[ibm["Attrition"] == 0][numerical_features].value_counts()
```

```python
ibm[ibm["Attrition"] == 0][numerical_features]
```

```python
ibm.groupby(['Attrition','MaritalStatus']).size()
```

```python
ibm.groupby(['Attrition','Gender']).size()
```

```python
ibm.groupby(['Attrition','Education']).size()
```

```python
ibm.groupby(['Attrition','BusinessTravel']).size()
```

```python
plt.figure(figsize=(20, 40))

for i, feature in enumerate(numerical_features, 1):
    plt.subplot(8, 3, i)
    ibm[ibm["Attrition"] == 0][feature].hist(bins=35, color='blue', label='Not Attrition', alpha=0.6)
    ibm[ibm["Attrition"] == 1][feature].hist(bins=35, color='red', label='Attrition', alpha=0.6)
    plt.legend()
    plt.xlabel(feature)
    plt.ylabel('count')
```

```python
plt.figure(figsize=(20, 15))

for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    ibm[ibm["Attrition"] == 0][feature].hist(bins=35, color='blue', label='Not Attrition', alpha=0.6)
    ibm[ibm["Attrition"] == 1][feature].hist(bins=35, color='red', label='Attrition', alpha=0.6)
    plt.legend()
    plt.xlabel(feature)
    plt.ylabel('count')
```
